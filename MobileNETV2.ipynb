```python
train_dir = os.path.join(path,"training_set")
test_dir = os.path.join(path,"test_set")
```

```python
data_gen_train = ImageDataGenerator(rescale = 1/255.)
data_gen_test = ImageDataGenerator(rescale = 1/255.)
```

```python
#class_mode = categorical
train_generator= data_gen_train.flow_from_directory(train_dir,target_size = (128,128),batch_size = 128,class_mode = "binary")
test_generator= data_gen_test.flow_from_directory(test_dir,target_size = (128,128),batch_size = 128,class_mode = "binary")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import imageio
import os

src_path = "/content/drive/MyDrive/dataset_lisa_jennie/training_set"
sub_train = os.listdir(src_path)
fig = plt.figure(figsize=(10,5))


for i in range(4):
    #path = os.path.join(src_path,sub_train[i])
    train_generator.filenames
    plt.subplot(240 + 1 + i)
    img = imageio.imread(os.path.join(src_path,train_generator.filenames[i]))
    plt.title(class_names[train_generator.classes[i]])
    plt.imshow(img, cmap=plt.get_cmap('gray'))
```

```python
fig = plt.figure(figsize=(10,5))
for i in range(4):
    #path = os.path.join(src_path,sub_train[i])
    train_generator.filenames
    plt.subplot(240 + 1 + i)
    img = imageio.imread(os.path.join(src_path,train_generator.filenames[140-i]))
    plt.title(class_names[train_generator.classes[140-i]])
    plt.imshow(img, cmap=plt.get_cmap('gray'))
```

```python
base_model = tf.keras.applications.MobileNetV2(input_shape=(128,128,3),include_top = False, weights = "imagenet") # include_top= false  fix weight and bias    head  
```

```python
base_model.summary()
```

```python
base_model_trainable = False  #fix weight and bias  model 
```

```python
average_pooling_layer = tf.keras.layers.GlobalAveragePooling2D()(base_model.output)
prediction_layer = tf.keras.layers.Dense(units=1,activation = 'sigmoid')(average_pooling_layer)
```

```python
model = tf.keras.models.Model(inputs=base_model.input, outputs = prediction_layer)
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001),loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history = model.fit(train_generator, batch_size=10,epochs=3,validation_data = test_generator)
```

```python
len(base_model.layers)
```

```python
base_model.trainable = True
```

```python
for layer in base_model.layers[:120]:  #layer 0-100  fix weight and bias 
  layer.trainable = False
```

```python
model.compile(optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.0001),loss="binary_crossentropy",metrics=["accuracy"])
```

```python
history = model.fit(train_generator,epochs=100,validation_data = test_generator)
```

```python
loss0, accuracy0 = model.evaluate(test_generator)
```

```python
print("initial loss: {:.2f}".format(loss0))
print("initial accuracy: {:.2f}".format(accuracy0))
```

```python
probabilities = model.predict_generator(generator=test_generator)
```

```python
y_true = test_generator.classes
```

```python
y_pred = (probabilities > 0.5).astype("int32")
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_true, y_pred)
print(cm)
```
