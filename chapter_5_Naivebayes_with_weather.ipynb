# Naive Bayes Classifier


```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
import os # use commandline
from google.colab import files
import io
uploaded = files.upload()
data  = pd.read_csv(io.BytesIO(uploaded['weather.csv']))
```

```python
data.info()
```

```python
data.columns
```

```python
data.head(14)
```

```python
X  = pd.DataFrame(data[['outlook', 'temperature', 'humidity', 'windy']])
y  = pd.DataFrame(data['class'])
```

```python
X
```

```python
from sklearn import preprocessing
X = X.apply(preprocessing.LabelEncoder().fit_transform)
```

```python
X
```

```python
print(X.info())
print(X.head())
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
# Creating Train and Test datasets
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 54, test_size = 0.33)
```

```python
X_test
```

```python
#Import Library of Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
```

```python
#Create a Gaussian Classifier
model = GaussianNB()
#clf1 = MultinomialNB()
#clf2 = BernoulliNB()
# Train the model using the training sets
model.fit(X_train, y_train)
#clf1.fit(X_train, y_train)
#clf2.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred= model.predict(X_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred=model.predict(X_test)))
```

```python
#Predict Output
predicted= model.predict([[0,1,0,0]] )
print(predicted)
```
